```python
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

# 1. The given dataset "Parkinson_Dataset_2" contains data from class 0 and 1. Split the dataset into 90:10 train -test ratio.


```python
data1 = pd.read_csv('Parkinson_Dataset_2.csv')
data1.describe()
```

```python
X = data1.drop('class', axis=1)
y = data1['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
```

```python
lda = LDA(n_components=1)
```

# 2. Perform LDA on the dataset on the 90% split

```python
lda.fit(X_train, y_train)
```

# 3. Test the accuracy on the 10% split. 

```python
lda.score(X_test, y_test)
```
